```python
import pandas as pd
import numpy as np
import wget
import requests
import os
import csv
```

```python
ciudad = "barranquilla"
tipo_archivo = "subjetivos"
resources_get = requests.get("http://comovamos.eokoe.com/api/3/action/package_show?id=barranquilla-subjetivos_anual")
resources = resources_get.json()["result"]["resources"]
```

```python
for resource in resources:
    wget.download(resource["url"],"data")
```

```python
def detect_typefile(string):
    if "diccionario" in string:
        return "diccionario"
    if "xlsx" in string:
        return "xlsx"
    return "data"
```

```python
def merge_data(file,data_dataframe):
    dataframe_file = pd.read_csv(file,delimiter=",", encoding="utf-8", dtype=np.string_ )
    dataframe_output = data_dataframe.append(dataframe_file)
    return dataframe_output
```

```python
def merge_diccionaries(file,diccionario_dataframe):
    dataframe_file = pd.read_csv(file,delimiter=",", encoding="utf-8", dtype=np.string_ )
    if len(diccionario_dataframe) == 0:
        return dataframe_file
    try:
        drop_rows = []
       
        for i, current_info in diccionario_dataframe.iterrows():
            current_variable = current_info["variable"]
            for j, current_file_info in dataframe_file.iterrows():
                current_file_variable = current_file_info["variable"]
                if current_variable == current_file_variable:
                    ano_min_diccionario = current_info["ano_min"]
                    ano_max_diccionario = current_info["ano_max"]
                    ano_min_file = current_file_info["ano_min"]
                    ano_max_file = current_file_info["ano_max"]
                    try:
                        if int(ano_min_file) < int(ano_min_diccionario):
                            diccionario_dataframe.loc[i,"ano_min"] = ano_min_file
                    except:
                        diccionario_dataframe.loc[i,"ano_min"] = ano_min_diccionario
                    try:
                        if int(ano_max_file) > int(ano_max_diccionario):
                            diccionario_dataframe.loc[i,"ano_max"] = ano_max_file
                    except:
                        diccionario_dataframe.loc[i,"ano_max"] = ano_max_diccionario
                    drop_rows.append(j)
                    
        dataframe_file = dataframe_file.drop(drop_rows, inplace=True)
        dataframe_output = diccionario_dataframe.append(dataframe_file)
    except:
        dataframe_output = diccionario_dataframe.append(dataframe_file)
    return dataframe_output
```

```python
path = "data/"
files = os.listdir(path)
data_dataframe = pd.DataFrame()
diccionario_dataframe = pd.DataFrame()
for file in files:
    filepath = path + file
    typefile = detect_typefile(filepath)
    if typefile != "xlsx":
        if typefile == "diccionario":
            diccionario_dataframe = merge_diccionaries(filepath,diccionario_dataframe)
        else:
            data_dataframe = merge_data(filepath,data_dataframe)

filename_data = path + ciudad + "_" + tipo_archivo + "_" + "lote.csv"
filename_diccionario = path + "diccionario_" + ciudad + "_" + tipo_archivo + "_" + "lote.csv"
            
            
data_dataframe.to_csv(filename_data,quoting=csv.QUOTE_ALL, encoding="utf-8", na_rep = np.nan, index = False)
diccionario_dataframe.to_csv(filename_diccionario,quoting=csv.QUOTE_ALL, encoding="utf-8", na_rep = np.nan, index = False)
```
